# Task: BANK TRANSACTION RECORD
A bank account records daily deposit transactions.
### Given data: 
### 5000, 12000, 7500, 3000, 9000

```python
# 1. Store the values in a Python list

transactions = [5000, 12000, 7500, 3000, 9000]
print("Transactions list:", transactions)
```

```python
# 2. Calculate and print the total amount deposited

total_deposited = sum(transactions)
print("Total amount deposited:", total_deposited)
```

```python
# 3. Find and print the highest deposit

highest_deposit = max(transactions)
print("Highest deposit:", highest_deposit)
```

```python
# 4. Find and print the lowest deposit

lowest_deposit = min(transactions)
print("Lowest deposit:", lowest_deposit)
```

```python
# 5. Add a new deposit of 6500 to the list

transactions.append(6500)
print("List after adding new deposit:", transactions)
```

```python
# 6. Remove the smallest deposit from the list

transactions.remove(min(transactions))
print("List after removing smallest deposit:", transactions)
```

```python
# 7. Print the final updated list of transactions

print("Final transactions list:", transactions)
```
